```python
#cleaning Population column
a = df.Population.tolist()
b = []
for i in a:
    i = i.replace("↗", "").replace(u"\xa0", "")
    i = int(i)
    b.append(i)
df.Population = b
```

```python
#opening my private yandex api key
import json

with open('C:/Users/Кирилл/api_key_yandex.json', mode='r') as key_file:
    key = json.loads(key_file.read())['key']
    
```

```python
#parcing latitude and longtitude data by yandex api (good works for russain adresses), 'c' is a list of boroughs in russian from cell 7.
import requests
latlon_data = {'Longitude':[], 'Latitute':[]}
for i in c:
    coor = requests.get('https://geocode-maps.yandex.ru/1.x/?geocode='+ i + '%2C+Район%2C+Москва&apikey=' + key+ '&lang=ru-RU&format=json&results=1')
    coor = coor.json()
    d = coor.get('response').get('GeoObjectCollection').get('featureMember')[0].get('GeoObject').get('Point').get('pos')
    Lon,Lat = d.split()
    latlon_data['Longitude'].append(Lon)
    latlon_data['Latitute'].append(Lat)
latlon_data
```

```python
#adding data to our dataframe
latlon_data = pd.DataFrame(latlon_data)
df = pd.concat([df, latlon_data], axis=1)
```

```python
with open('C:/Users/Кирилл/id_foursquare.json', mode='r') as id_file:
    CLIENT_ID = json.loads(id_file.read())['key'] 
with open('C:/Users/Кирилл/secret_foursquare.json', mode='r') as secret_file:
    CLIENT_SECRET = json.loads(secret_file.read())['key']
VERSION = '20180605' # Foursquare API version
LIMIT = 2000 # A default Foursquare API limit value
```

```python
sheeeeet_data = {'Number of places':[]}
```

```python
#parcing info about places number
for i in range(146):
    neighborhood_latitude = df.loc[i, 'Latitute'] # neighborhood latitude value
    neighborhood_longitude = df.loc[i, 'Longitude'] # neighborhood longitude value
    radius = df.loc[i, 'Area']*1000**0.5
    url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
        CLIENT_ID, 
        CLIENT_SECRET, 
        VERSION, 
        neighborhood_latitude, 
        neighborhood_longitude, 
        radius, 
        LIMIT)
    results = requests.get(url).json()
    res = results.get('response').get('totalResults')
    sheeeeet_data['Number of places'].append(res)
```

```python
#adding data to our dataframe
sheeeeet_data = pd.DataFrame(sheeeeet_data)
df = pd.concat([df, sheeeeet_data], axis=1)
df
```
